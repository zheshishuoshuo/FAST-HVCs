```python
import numpy as np
import matplotlib.pyplot as plt

# 生成随机的斜线段
slope = 0.5
intercept = 5
x1 = np.arange(0, 10, 0.1)
y1 = slope*x1 + intercept
y1 += np.random.normal(0, 0.5, size=y1.shape)

# 生成随机的平行线段
x2 = np.arange(10, 20, 0.1)
y2 = np.ones_like(x2) * intercept
y2 += np.random.normal(0, 0.5, size=y2.shape)

# 将两个线段的首尾相连
x = np.concatenate([x1, x2[1:]])
y = np.concatenate([y1, y2[1:]])

# 绘制图形
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(x, y, '-o', alpha=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Staircase Curve')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成随机的斜线段
slope = 0.3
intercept = 1.5
x1 = np.arange(0, 5, 0.1)
y1 = np.ones_like(x1) * intercept
y1 += np.random.normal(0, 0.01, size=y1.shape)

# 生成随机的平行线段
x2 = np.arange(5, 10, 0.1)
y2 = slope*(x2-5) + intercept
y2 += np.random.normal(0, 0.01, size=y2.shape)

# 将两个线段拼接在一起
x = np.concatenate([x1, x2])
y = np.concatenate([y1, y2])

# 绘制图形
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(x, y, '-o', alpha=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Staircase Curve')
plt.show()
```
